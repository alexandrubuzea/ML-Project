```python
import json
import pandas as pd

results_list = []
with open('results_linear.json', 'r') as fin:
    results_list = json.load(fin)

final_results = {
    "hidden_layer_size": [],
    "learning_rate" : [],
    "weight_decay" : [],
    "use_adam": [],
    "learning_rate_decay": [],
    "epochs" : [],
    "batch_size" : [],
    "hidden_layers_count" : [],
    "train_loss" : [],
    "test_loss" : [],
    "train_acc" : [],
    "test_acc" : []
}

for entry in results_list:
    for key in entry:
        if type(entry[key]) is list:
            final_results[key].append(entry[key][-1])
        else:
            final_results[key].append(entry[key])

final_results = pd.DataFrame(final_results)

final_results.head()
```

```python
best_result = results_list[41]

import matplotlib.pyplot as plt

plt.figure(figsize=(15,10))
plt.title('Evolution of loss for patients dataset')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.plot(list(range(len(best_result['train_loss']))), best_result['train_loss'], label="Train loss")
plt.plot(list(range(len(best_result['train_loss']))), best_result['test_loss'], label="Test loss")
plt.legend(loc='upper right', title='Loss')

plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.title('Evolution of accuracy for patients dataset')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.plot(list(range(len(best_result['train_loss']))), best_result['train_acc'], label="Train acc")
plt.plot(list(range(len(best_result['train_loss']))), best_result['test_acc'], label="Test acc")
plt.legend(loc='upper right', title='Loss')
```

```python
final_results.to_csv('results_patients.csv', index=False)
```
