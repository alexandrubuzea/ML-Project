```python
def plot_time_series(data, type):
    plt.figure(figsize=(15,10))
    plt.title('Time series for class ' + type)
    plt.xlabel('Time')
    plt.ylabel('ECG')
    plt.plot(list(range(len(data))), list(data))
    plt.show()
```

```python
plot_time_series(normal_df.iloc[0, 0:-1], "normal")
plot_time_series(abnormal_df.iloc[0, 0:-1], "abnormal")

plot_time_series(normal_df.mean()[0:-1], "normal (mean)")
plot_time_series(abnormal_df.mean()[0:-1], "abnormal (mean)")
plot_time_series(normal_df.std()[0:-1], "normal (std)")
plot_time_series(abnormal_df.std()[0:-1], "abnormal (std)")
```

```python
from numpy.fft import fft

def plot_freq_series(data, type):
    plt.figure(figsize=(15,10))
    plt.title('Frequency series for class ' + type)
    plt.xlabel('Frequency')
    plt.ylabel('ECG')

    t = np.arange(len(data))
    sp = np.fft.fft(list(data))
    freq = np.fft.fftfreq(t.shape[-1])
    
    plt.plot(freq, abs(sp))
    plt.show()
```

```python
plot_freq_series(normal_df.mean()[0:-1], "normal (frequency)")
plot_freq_series(abnormal_df.mean()[0:-1], "abnormal (frequency)")
```

```python
# implement neural network for the patients's dataset

patients_df = pd.read_csv('patients_data.csv')

patients_df.head()
```

```python
X = patients_df.iloc[:, 0:-1].values
y = patients_df.iloc[:, -1].values

print(X.shape)
print(y.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X = scaler.fit_transform(X)
```

```python
y = list(map(lambda s: int(s[1:]), y))
```

```python
set(y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True, random_state=42)
```

```python
import torch
import torch.nn as nn
from torch.autograd import Variable
```

```python
HIDDEN_LINEAR = 500
LR_LINEAR = 0.7
WEIGHT_DECAY_LINEAR = 0.1
USE_ADAM_LINEAR = False # True if uses Adam optimizer, False if uses SGD with momentum and LR scheduler
LR_DECAY_LINEAR = 0.95
EPOCHS_LINEAR = 100
BATCH_SIZE_LINEAR = 64
NUM_HIDDEN_LINEAR = 4
```

```python
TRAIN_LOSS = 'train_loss'
TEST_LOSS = 'test_loss'
TRAIN_ACC = 'train_acc'
TEST_ACC = 'test_acc'
```

```python
results = {
    "hidden_layer_size": HIDDEN_LINEAR,
    "learning_rate": LR_LINEAR,
    "weight_decay": WEIGHT_DECAY_LINEAR,
    "use_adam": USE_ADAM_LINEAR,
    "learning_rate_decay": LR_DECAY_LINEAR,
    "epochs": EPOCHS_LINEAR,
    "batch_size": BATCH_SIZE_LINEAR,
    "hidden_layers_count" : NUM_HIDDEN_LINEAR,
    TRAIN_LOSS: [],
    TEST_LOSS: [],
    TRAIN_ACC: [],
    TEST_ACC: []
}
```

```python
class CustomDataset(torch.utils.data.Dataset):
    def __init__(self, data, labels, transform=None):
        self.data = data
        self.labels = labels
        self.transform = transform

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        sample = self.data[idx]
        label = self.labels[idx]

        sample = torch.tensor(sample, dtype=torch.float32)
        label = torch.tensor(label, dtype=torch.long)

        if self.transform:
            sample = self.transform(sample)

        return sample, label
```

```python
train_loader = torch.utils.data.DataLoader(dataset=CustomDataset(data=X_train, labels=y_train), batch_size=BATCH_SIZE_LINEAR, shuffle=True)
test_loader = torch.utils.data.DataLoader(CustomDataset(data=X_test, labels=y_test), batch_size=BATCH_SIZE_LINEAR, shuffle=False)
```

```python
class PatientsNN(nn.Module):
    def __init__(self, in_size, out_size):
        super(PatientsNN, self).__init__()

        self.relu = nn.ReLU()

        self.linear_initial = nn.Linear(in_size, HIDDEN_LINEAR)
        self.linear_hidden = [nn.Linear(HIDDEN_LINEAR, HIDDEN_LINEAR) for _ in range(NUM_HIDDEN_LINEAR)]
        self.linear_final = nn.Linear(HIDDEN_LINEAR, out_size)

    def forward(self, x):
        y = self.linear_initial(x)
        y = self.relu(y)

        for layer in self.linear_hidden:
            y = layer(y)
            y = self.relu(y)
        
        y = self.linear_final(y)

        return y

    def name(self):
        return "PatientsNN"
```

```python
in_size = X.shape[1]
out_size = len(set(y))
model = PatientsNN(in_size, out_size)
```

```python
print(out_size)
```

```python
use_cuda = torch.cuda.is_available()
if use_cuda:
    model = model.cuda()
```

```python
adam_optimizer = torch.optim.Adam(params=model.parameters(), lr=LR_LINEAR, weight_decay=WEIGHT_DECAY_LINEAR)
sgd_optimizer = torch.optim.SGD(params=model.parameters(), lr=LR_LINEAR, momentum=0.8)

lr_decay = lambda epoch: LR_DECAY_LINEAR ** epoch
scheduler = torch.optim.lr_scheduler.LambdaLR(optimizer=sgd_optimizer, lr_lambda=lr_decay)
```

```python
loss_function = nn.CrossEntropyLoss()
```

```python
optimizer = adam_optimizer if USE_ADAM_LINEAR else sgd_optimizer

for epoch in range(EPOCHS_LINEAR):
    model.train()

    avg_loss = 0
    
    expected = 0
    correct = 0

    for index, (x, target) in enumerate(train_loader):
        optimizer.zero_grad()

        if use_cuda:
            x = x.cuda()
            target = target.cuda()

        x = Variable(x)
        target = Variable(target)

        outputs = model(x)
        loss = loss_function(outputs, target)

        loss.backward()

        optimizer.step()

        if USE_ADAM_LINEAR:
            scheduler.step()
        
        avg_loss = 0.9 * avg_loss + 0.1 * loss.item()

        predictions = torch.argmax(outputs, axis=1)
        
        correct += torch.sum(predictions == target).numpy()
        expected += len(target)

    results[TRAIN_LOSS].append(avg_loss)
    results[TRAIN_ACC].append(correct * 1.0 / expected)
    avg_loss = 0

    correct = 0
    expected = 0

    model.eval()

    for index, (x, target) in enumerate(test_loader):

        if use_cuda:
            x = x.cuda()
            target = target.cuda()

        x = Variable(x)
        target = Variable(target)

        outputs = model(x)
        loss = loss_function(outputs, target)

        avg_loss = 0.9 * avg_loss + 0.1 * loss.item()

        predictions = torch.argmax(outputs, axis=1)
        
        correct += torch.sum(predictions == target).numpy()
        expected += len(target)

    results[TEST_LOSS].append(avg_loss)
    results[TEST_ACC].append(correct * 1.0 / expected)

```

```python
final_test_loader = torch.utils.data.DataLoader(CustomDataset(data=X_test, labels=y_test), batch_size=len(y_test), shuffle=False)

model.eval()

outputs = model(torch.Tensor(X_test))
predictions = torch.argmax(outputs, axis=1)

from sklearn.metrics import confusion_matrix, accuracy_score

plt.figure(figsize=(10, 10))
sns.heatmap(confusion_matrix(y_test, predictions), annot=True, cmap='Blues', fmt='g', cbar=False)
plt.title('Confusion Matrix for Patients')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

```python
# classes = list(map(lambda x: "D" + str(x), list(range(7))))

from sklearn.metrics import f1_score, precision_score, recall_score

for label in range(7):
    print(f"Label {label}: P = {precision_score(y_test, predictions, average='macro', labels=[label])}, R = {recall_score(y_test, predictions, average='macro',  labels=[label])}, F1 = {f1_score(y_test, predictions, average='macro',  labels=[label])}")
```

```python
# Part 2: Time series analysis

X = ptb_df.iloc[:, 0:-1].values
y = ptb_df.iloc[:, -1].values

scaler = StandardScaler()

X = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True, random_state=42)

print(X_train.shape)
print(X_test.shape)
print(set(y))
```

```python
HIDDEN_CONV = 500
LR_CONV = 0.7
WEIGHT_DECAY_CONV = 10.0
USE_ADAM_CONV = False # True if uses Adam optimizer, False if uses SGD with momentum and LR scheduler
LR_DECAY_CONV = 0.95
EPOCHS_CONV = 100
BATCH_SIZE_CONV = 128
NUM_HIDDEN_CONV = 4
OUT_CHANNELS = 6
KERNEL_SIZE = 3
OUT_CHANNELS2 = 16
KERNEL_SIZE2 = 3
```

```python
results = {
    "hidden_layer_size": HIDDEN_CONV,
    "learning_rate": LR_CONV,
    "weight_decay": WEIGHT_DECAY_CONV,
    "use_adam": USE_ADAM_CONV,
    "learning_rate_decay": LR_DECAY_CONV,
    "epochs": EPOCHS_CONV,
    "batch_size": BATCH_SIZE_CONV,
    "hidden_layers_count" : NUM_HIDDEN_CONV,
    "kernel_size": KERNEL_SIZE,
    "kernel_size2": KERNEL_SIZE2,
    "out_channels": OUT_CHANNELS,
    "out_channels2": OUT_CHANNELS2,
    TRAIN_LOSS: [],
    TEST_LOSS: [],
    TRAIN_ACC: [],
    TEST_ACC: []
}
```

```python
train_loader = torch.utils.data.DataLoader(dataset=CustomDataset(data=X_train, labels=y_train), batch_size=BATCH_SIZE_CONV, shuffle=True)
test_loader = torch.utils.data.DataLoader(CustomDataset(data=X_test, labels=y_test), batch_size=BATCH_SIZE_CONV, shuffle=False)
```

```python
from math import floor

class EcgNN(nn.Module):
    def __init__(self, in_size, out_size):
        super(EcgNN, self).__init__()

        self.relu = nn.ReLU()

        self.conv1 = nn.Conv1d(in_channels=1, out_channels=OUT_CHANNELS, kernel_size=KERNEL_SIZE, padding=int(((KERNEL_SIZE - 1) / 2)))
        self.conv2 = nn.Conv1d(in_channels=OUT_CHANNELS, out_channels=OUT_CHANNELS2, kernel_size=KERNEL_SIZE2, padding=int(((KERNEL_SIZE2 - 1) / 2)))
        self.pool = nn.MaxPool1d(kernel_size=2)
        self.linear_initial = nn.Linear(floor(floor(in_size / 2) / 2) * OUT_CHANNELS2, HIDDEN_CONV)
        self.linear_hidden = [nn.Linear(HIDDEN_CONV, HIDDEN_CONV) for _ in range(NUM_HIDDEN_CONV)]
        self.linear_final = nn.Linear(HIDDEN_CONV, out_size)

    def forward(self, x):
        x = x.view(x.size(0), 1, x.size(1))
        # print(x.shape)
        y = self.conv1(x)
        y = self.relu(y)
        y = self.pool(y)

        y = self.conv2(y)
        y = self.relu(y)
        y = self.pool(y)

        y = torch.flatten(y, 1)
        y = self.linear_initial(y)
        y = self.relu(y)

        for layer in self.linear_hidden:
            y = layer(y)
            y = self.relu(y)
        
        y = self.linear_final(y)

        return y

    def name(self):
        return "EcgNN"
```

```python
in_size = X.shape[1]
out_size = len(set(y))
model = EcgNN(in_size, out_size)
```

```python
adam_optimizer = torch.optim.Adam(params=model.parameters(), lr=LR_CONV, weight_decay=WEIGHT_DECAY_CONV)
sgd_optimizer = torch.optim.SGD(params=model.parameters(), lr=LR_CONV, momentum=0.8)

lr_decay = lambda epoch: LR_CONV ** epoch
scheduler = torch.optim.lr_scheduler.LambdaLR(optimizer=sgd_optimizer, lr_lambda=lr_decay)
```

```python
loss_function = nn.CrossEntropyLoss()
```

```python
optimizer = adam_optimizer if USE_ADAM_CONV else sgd_optimizer

for epoch in range(EPOCHS_CONV):
    model.train()

    avg_loss = 0
    
    expected = 0
    correct = 0

    for index, (x, target) in enumerate(train_loader):
        optimizer.zero_grad()

        if use_cuda:
            x = x.cuda()
            target = target.cuda()

        x = Variable(x)
        target = Variable(target)

        outputs = model(x)
        loss = loss_function(outputs, target)

        loss.backward()

        optimizer.step()

        if USE_ADAM_CONV:
            scheduler.step()
        
        avg_loss = 0.9 * avg_loss + 0.1 * loss.item()

        predictions = torch.argmax(outputs, axis=1)
        
        correct += torch.sum(predictions == target).numpy()
        expected += len(target)

    results[TRAIN_LOSS].append(avg_loss)
    results[TRAIN_ACC].append(correct * 1.0 / expected)
    avg_loss = 0

    correct = 0
    expected = 0

    model.eval()

    for index, (x, target) in enumerate(test_loader):

        if use_cuda:
            x = x.cuda()
            target = target.cuda()

        x = Variable(x)
        target = Variable(target)

        outputs = model(x)
        loss = loss_function(outputs, target)

        avg_loss = 0.9 * avg_loss + 0.1 * loss.item()

        predictions = torch.argmax(outputs, axis=1)
        
        correct += torch.sum(predictions == target).numpy()
        expected += len(target)

    results[TEST_LOSS].append(avg_loss)
    results[TEST_ACC].append(correct * 1.0 / expected)

```

```python
final_test_loader = torch.utils.data.DataLoader(CustomDataset(data=X_test, labels=y_test), batch_size=len(y_test), shuffle=False)

model.eval()

outputs = model(torch.Tensor(X_test))
predictions = torch.argmax(outputs, axis=1)

from sklearn.metrics import confusion_matrix, accuracy_score

plt.figure(figsize=(10, 10))
sns.heatmap(confusion_matrix(y_test, predictions), annot=True, cmap='Blues', fmt='g', cbar=False)
plt.title('Confusion Matrix for PTB')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```
